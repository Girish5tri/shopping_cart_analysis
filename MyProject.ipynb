```python
#In [1]: Importing Libraries
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("dark")
import matplotlib
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
#In [2]: Reading Data Sets
data = pd.read_csv('/content/Market_Basket_Optimisation.csv',header=None)
```

```python
#In [6]: Append the dataset values into an array
trans=[]
for i in range(len(data)):
 trans.append([str(data.values[i,j]) for j in range(0,20)])
trans=np.array(trans)
print(trans.shape)
```

```python
#In [7]:
trans
```

```python
#In [8]:
t=TransactionEncoder()
data=t.fit_transform(trans)
data=pd.DataFrame(data,columns=t.columns_,dtype=int)
data.shape
```

```python
#In [9]:   Drop the ‘nan’ values
data.drop('nan',axis=1,inplace=True)
```

```python
#In [10]:
data.shape
'nan' in data.columns
```

```python
#In [11]:
data.head()
```

```python
#In [12]: Exploratory data analysis (EDA) of top 20 selling items
r=data.sum(axis=0).sort_values(ascending=False)[:20]
plt.figure(figsize=(20,10))
s=sns.barplot(x=r.index,y=r.values)
s.set_xticklabels(s.get_xticklabels(), rotation=90)
plt.show()
```

```python
#In [13]: Tree map of the 20 most selling items
#pip install squarify
!pip install squarify -qqq
import squarify
#In [14]:
my_values=r.values
cmap = matplotlib.cm.Blues
mini=min(my_values)
maxi=max(my_values)
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in my_values]
plt.figure(figsize=(10,10))
squarify.plot(sizes=r.values, label=r.index, alpha=.7,color=colors)
plt.title("Tree map of top 20 items")
plt.axis('off')
plt.show()
```

```python
#In [15]: Making frequent items lists using Apriori where minimum support = 0.05
freq_items=apriori(data,min_support=0.05,use_colnames=True)
```

```python
#In [16]:
freq_items
```

```python
#In [17]: Association rule1 of the freq_items where min_threshold for lift = 1.3
res=association_rules(freq_items,metric="lift",min_threshold=1.3)
```

```python
#In[18]:
res
```

```python
#In [19]: Association rule2 of the freq_items where min_threshold for confidence = 0.3
res1=association_rules(freq_items,metric="confidence",min_threshold=0.3)
res1
```

```python
#In [20]: Association rule2 of the freq_items where some selected attributes are displayed
res2 = res[['antecedents','consequents','support','confidence','lift']]
res2
```

```python
#In [21]:
res3= res[res['confidence']>=0.1]
res3
```

```python
#In [22]: Listing frequent_itemsets where minimum support = 0.05
frequent_itemsets = apriori(data, min_support = 0.05, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
#In[30]: Comparison of RunTime between Apriori and FPGrowth using Barplot and Lineplot
import time
l=[0.01,0.02]
t=[]
for i in l:
 t1=time.time()
 apriori(data,min_support=i,use_colnames=True)
 t2=time.time()
 t.append((t2-t1)*1000)
```

```python
#In[31]:
l=[0.01,0.02]
f=[]
for i in l:
 t1=time.time()
 fpgrowth(data,min_support=i,use_colnames=True)
 t2=time.time()
 f.append((t2-t1)*1000)
```

```python
#In[32]: Lineplot
sns.lineplot(x=l,y=f,label="fpgrowth")
sns.lineplot(x=l,y=t,label="apriori")
plt.xlabel("Min_support Threshold")
plt.ylabel("Run Time in ms")
plt.show()
```

```python
#In[33]: Barplot
sns.barplot(x=l,y=f,label="fpgrowth")
sns.barplot(x=l,y=t,label="apriori")
plt.xlabel("Min_support Threshold")
plt.ylabel("Run Time in ms")
plt.show()
```
